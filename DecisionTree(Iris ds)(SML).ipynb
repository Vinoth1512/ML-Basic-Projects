```python
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
data=sns.load_dataset('iris')
data.head()
```

```python
x=data.iloc[:,0:4]
x.head()
```

```python
y=data.iloc[:,-1]
y.head()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=1/3,random_state=70)
model=DecisionTreeClassifier()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
ytest.head()
```

```python
ypred[0:5]
```

```python
print("Accuracy Score =",(accuracy_score(ytest,ypred)*100),"%")

```

```python
xplt=data["sepal_length"]
yplt=data["species"]
plt.scatter(xplt,yplt,color='r',label="Sepal Length")

xplt=data["sepal_width"]
yplt=data["species"]
plt.scatter(xplt,yplt,color='b',label="Sepal width")

xplt=data["petal_length"]
yplt=data["species"]
plt.scatter(xplt,yplt,color='m',label="Petal Length")

xplt=data["petal_width"]
yplt=data["species"]
plt.scatter(xplt,yplt,color='g',label="Petal Width")

plt.legend(prop={'size':8})
plt.title("IRIS CLASSIFICATION")
plt.show()
```

```python
sns.jointplot(x="sepal_length",y="sepal_width",data=data,hue="species")
plt.show()
```
